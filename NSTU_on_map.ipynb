# Circle marker on map(Bangladesh)

```python
import folium
```

```python
m= folium.Map(location=[22.7908, 91.0990],zoom_start=15)
folium.CircleMarker(
   location=[22.7908, 91.0990],
   radius=80,
   popup="Noakhali Science and Technology University (NSTU)",
   color="Red",
   fill=True,
   fill_color="orange",
).add_to(m)
m
```
