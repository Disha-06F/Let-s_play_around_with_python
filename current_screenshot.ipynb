### A library for automating GUI interactions, including taking Screenshots.
<span style="color:green">PLI(pillow):</span>
### A library for handling images(used to open the screenshot)
<span style="color:red">myScreenshot = pyautogui.screenshot()</span>
### Captures the entire screen and stores the image in myScreenshot.
<span style="color:red">myScreenshot.save(r'clcoding.png')</span>
### Saves the screenshot as "clcoding.png" in the current working directory.
### The r (raw string) ensures that backslashes in file paths are handled correctly.
<span style="color:red">Image.open(r'clcoding.png')</span>
### Opens the saved image using PIL.Image.


```python
import pyautogui
from PIL import Image

myScreenshot = pyautogui.screenshot()
myScreenshot.save(r'clcoding.png')
Image.open(r'clcoding.png')
```
